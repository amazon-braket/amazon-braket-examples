# Using the CUDA-Q Applications Hub

Welcome! This notebook provides a quick and easy way to pull the NVIDIA CUDA-Q Applications Hub directly into your Amazon Braket environment.

Designed for researchers and developers, these notebooks feature industry-driven use cases that demonstrate CUDA-Q's ability to seamlessly integrate GPU acceleration with quantum processors and scale complex hybrid algorithms.

You can browse these examples in the [CUDA-Q Applications Hub](https://nvidia.github.io/cuda-quantum/latest/using/applications.html).


The command below will clone the application library notebooks into a local folder called `cuda-q-applications` so that you can explore and run them directly within this notebook environment. (If the folder already exists locally, it will be moved to a backup folder.)

<div class="alert alert-block alert-warning">
    <b>Note:</b> Running the cell below will download content that is provided by NVIDIA. Run this cell to clone the library:

```python
![ -d cuda-q-applications ] && mv cuda-q-applications cuda-q-applications-backup-$(date +%Y%m%d-%H%M%S); git clone --filter=blob:none --sparse --depth 1 https://github.com/NVIDIA/cuda-quantum.git cuda-q-applications && cd cuda-q-applications && git sparse-checkout set docs/sphinx/applications/python && rm -rf .git && cd .. && mv cuda-q-applications cuda-q-applications-temp && mv cuda-q-applications-temp/docs/sphinx/applications/python cuda-q-applications && rm -rf cuda-q-applications-temp && echo "" && echo "âœ… Success! The CUDA-Q Applications Hub has been cloned to the \`cuda-q-applications\` folder"
```

You can now navigate into the `cuda-q-applications` folder to explore the library.
