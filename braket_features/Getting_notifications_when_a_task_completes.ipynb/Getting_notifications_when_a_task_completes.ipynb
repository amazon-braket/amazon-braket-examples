# Getting notifications when a task completes

This notebook walks you through how to set up notifications for completion of Amazon Braket tasks through the Amazon Simple Notification Service (SNS). Active notifications may be useful in situations where a large wait time is expected, such as when a submitted task is large and takes a while to complete, or when a task is submitted to a device outside of its availability window. In such a setting, a user may not want to wait for the task to complete, and would prefer to move forward and receive an alert once the task is complete.

## The workflow

Amazon Braket tasks leverage Amazon S3 as an intermediate storage device. This allows you to leverage the built-in event architecture of S3 to generate active push notifications. All you need to do is subscribe an SMS capable endpoint (e.g. an email address or a cell phone number) to an SNS Topic, and tie that SNS Topic to the S3 bucket of interest.

# Instructions

### Optional Step: Create a KMS customer master key (CMK)

Security is a priority at AWS, and as such, nearly all the existing AWS Services integrate with AWS Key Management Service (KMS) to allow for data to be encrypted both in transmission and at rest. S3 and SNS are both integrated with KMS, so if you would like to leverage this, a quick summary is provided here. First, create a custom key via the KMS console via the following instructions:

https://docs.aws.amazon.com/kms/latest/developerguide/create-keys.html#create-symmetric-cmk

**NOTE: When creating the key, you don't need to explicitly allow any IAM users, roles, or services to have access to the key.**

After the key is created, you need to edit the default Key Policy to include both services as Principals which can encrypt and decrypt data. Add the following permission within the "Statement" list (remember to replace the `<CMK ARN>` into the ARN of the key). 

```        
{
    "Sid": "Allow access for S3 and SNS (as Service Principals)",
    "Effect": "Allow",
    "Principal": {
        "Service": [
            "s3.amazonaws.com",
            "sns.amazonaws.com"
        ]
    },
    "Action": [
        "kms:GenerateDataKey*",
        "kms:Decrypt"
    ],
    "Resource": "<CMK ARN>"
}
```

### Step 1: Create Appropriate S3 Bucket

If you haven't already done so, create an S3 bucket for the Braket task results to land in. To accommodate certain IAM permissions, Braket expects the bucket to be named along the following lines:

`amazon-braket-<bucket name>`

All other tutorials expect a bucket name of this form. 

### Step 2: Create an SNS Topic

Amazon Simple Notification Service (SNS) acts as the event broker that receives notifications from S3 (along a given Topic) and routes them out to any subscriber, e.g., an email address. You need to create that SNS Topic to serve as that central communcations channel. See the following instructions:

https://docs.aws.amazon.com/sns/latest/dg/sns-tutorial-create-topic.html#create-topic-aws-console

**NOTE: If you created a CMK in the optional step, make sure you enable encryption on the SNS Topic, and select the CMK you created from the dropdown menu.**

You should already have an S3 bucket `amazon-braket-<bucket name>` created for use in this flow. You need to add permissions to the SNS Topic to allow the bucket to send events to it. To do so, edit the Topic's "Access Policy" to paste the following permission (under the "Statement" list). You need to replace the contents in <>. 

```
{
    "Sid": "allow-S3-access",
    "Effect": "Allow",
    "Principal": {
        "AWS": "*"
    },
    "Action": "SNS:Publish",
    "Resource": "arn:aws:sns:<region>:<account_id>:<sns_topic_name>",
    "Condition": {
        "StringEquals": {
            "aws:SourceAccount": "<account_id>"
        },
        "ArnLike": {
            "aws:SourceArn": "arn:aws:s3:*:*:amazon-braket-<bucket name>"
        }
    }
}
```

### Step 3: Subscribe an endpoint to the SNS Topic

Follow the instructions here to subscribe an endpoint to an SNS Topic:

https://docs.aws.amazon.com/sns/latest/dg/sns-create-subscribe-endpoint-to-topic.html

SNS > Subscriptions > Create subscription > choose the Topic ARN from the dropdown menu > choose the protocol from the dropdown menu (SMS for cell phone number and Email for email address) > fill in the details of the endpoint

**NOTE: SNS is not a free service. Check the pricing [here](https://aws.amazon.com/sns/pricing/). Sending text messages to non-US cell phone numbers might be expensive. On the other hand, the default spending limit for SMS messages is 1.00 USD per month. You can [request](https://aws.amazon.com/premiumsupport/knowledge-center/sns-sms-spending-limit-increase/) a limit increase through the support center.**

Once this step is completed, you should see the endpoint listed under the SNS > Subscriptions tab in the console. To test functionality, you can go to the subscribed Topic and publish a test message via "Publish message". If the endpoint is properly subscribed, you will receive a  message immediately with the message title and text you provided. If you do not receive a text, double check the instructions in this step.

### Step 4: Create an S3 Event and tie it to the SNS Topic

Follow the instructions provided here to create a new S3 Event:

https://docs.aws.amazon.com/AmazonS3/latest/user-guide/enable-event-notifications.html#enable-event-notifications-how-to

Open the Amazon S3 console > choose the bucket `amazon-braket-<bucket name>` > Properties > choose Create event notification under Event notifications > Enter event name, prefix, and suffix > choose event type > choose SNS topic as the destination > select the SNS topic created in Step 2 from the dropdown menu.

**NOTE 1: It is recommended to filter the S3 events by prefix to prevent unnecessary notifications. For the purpose of this tutorial, you can add a prefix filter on `sns-testing/` , which is used as the destination in the following cell for Braket task outputs.**

**NOTE 2: If the SNS Topic ARN cannot be found in the dropdown menu, it is likely that the appropriate permissions were not added to the Topic Access Policy - review Step 2.**

### Step 5: Test the setup with a Braket task

By this point, the pipeline is configured and ready to test. To do so, run the following code to kick off a Braket task. Once the task state registers as "COMPLETED", it should only take a few seconds for you to receive a message notifying a file has been added to the S3 bucket, indicating the task is done. 

````python
from braket.circuits import Circuit
from braket.aws import AwsDevice

n_qubits = 10
n_shots = 100

# set up S3 bucket (where results are stored)
my_bucket = 'amazon-braket-<bucket name>'  # enter your bucket here
my_prefix = 'sns-testing'               # enter your prefix here
s3_folder = (my_bucket, my_prefix)

device = AwsDevice('arn:aws:braket:::device/quantum-simulator/amazon/sv1')
# ##

def ghz_circuit(n_qubits: int) -> Circuit:
    """
    Function to return simple GHZ circuit ansatz. Assumes all qubits in range(0, n_qubits-1)
    are entangled.

    :param int n_qubits: number of qubits
    :return: Constructed GHZ circuit
    :rtype: Circuit
    """

    circuit = Circuit()                          # instantiate circuit object
    circuit.h(0)                                 # add Hadamard gate on first qubit

    for ii in range(0, n_qubits-1):
        circuit.cnot(control=ii, target=ii+1)    # apply series of CNOT gates
    return circuit


# Define circuit
ghz = ghz_circuit(n_qubits)

# Kick off single task execution
task = device.run(ghz, s3_folder, n_shots, poll_timeout_seconds=3600)
````
